#### Only 159, so there are ~1085-159 = 926 images with at least one annotations score of 0.001 but also with annotation/s with higher score. 
Where does this 0.001 comes from?
Apperantly it's the minimun score given in the openpifpaf/annotations.py file.

    def json_data(self):
        """Data ready for json dump."""

        # avoid visible keypoints becoming invisible due to rounding
        v_mask = self.data[:, 2] > 0.0
        keypoints = np.copy(self.data)
        keypoints[v_mask, 2] = np.maximum(0.01, keypoints[v_mask, 2])
        keypoints = np.around(keypoints.astype(np.float64), 2)

        # convert to float64 before rounding because otherwise extra digits
        # will be added when converting to Python type
        data = {
            'keypoints': keypoints.reshape(-1).tolist(),
            'bbox': [round(float(c), 2) for c in self.bbox()],
            'score': max(0.001, round(self.score(), 3)),
            'category_id': self.category_id,
        }

        id_ = getattr(self, 'id_', None)
        if id_:
            data['id_'] = id_

        return data

### Choose random image per score thresh, plot image and image with keypoints prediction

```python
thresh_ranges = list(zip(np.round(np.arange(0.1, 1, 0.1), 4), np.round(np.arange(0.2, 1.1, 0.1), 4)))
thresh_ranges_str = [str(t) for t in thresh_ranges]
@interact(thresh_range=thresh_ranges_str)
def predict_and_plot(thresh_range_str=thresh_ranges_str):
    thresh_range = tuple(map(float, thresh_range_str.strip('()').split(', ')))
    min_thresh_score = float(thresh_range[0])
    max_thresh_score = float(thresh_range[1])
    image_count = 4
    round_digits = 4
    chosen_images = dict
    images_ids = df_ann_scores_description.loc[(df_ann_scores_description['min'] > min_thresh_score)&(df_ann_scores_description['max'] <= max_thresh_score)].index.values
    if images_ids.shape[0] > 0:
        chosen_images = np.random.choice(images_ids, image_count, replace=False)    
    for image_id in chosen_images:
        print('thresh range: {thresh_range}, image_id: {image_id}'.format(thresh_range=thresh_range,
                                                                          image_id=image_id))
        image_id_length = len(str(image_id))
        image_file_partial = (total_image_file_name_length-image_id_length) * '0' + str(image_id)
        image_file_full = image_file_partial + '.jpg'
        image_path = os.path.join(val_images_directory, image_file_full)
        pred_output = str(pathlib.Path('notebooks', 'pred', image_file_partial + '_pred'))
        model_path = os.path.join(curr_directory, model_name)
        cmd = 'cd .. && python -m openpifpaf.predict {image_path} --image-output {pred_output} --checkpoint {model_path}'.format(image_path=image_path,
                                                                                                                                 pred_output=pred_output,
                                                                                                                                         model_path=model_path)
        print(cmd)
        return_value = os.system(cmd)
        assert (return_value == 0)

        img = Image.open(os.path.join('..', pred_output + '.png'))
        display(img)
        
```
